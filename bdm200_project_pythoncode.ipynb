```python
import numpy as np
import pandas as pd

# Loading the news vectors from the Excel file
newsvectors = pd.read_excel("C:/Users/TANVI/Downloads/news_vectors.xlsx", header=None)

# Converting the DataFrame to a numpy array
news_vectors = newsvectors.to_numpy()

# Calculating the similarity matrix
numberarticles, vecdim = news_vectors.shape
similarity_matrix = np.zeros((numberarticles, numberarticles))

for i in range(numberarticles):
    for j in range(numberarticles):
        similarity = np.dot(news_vectors[i], news_vectors[j])
        similarity_matrix[i, j] = similarity

# Find the most similar article for each article (excluding self-similarity)
sim_articles = np.argmax(similarity_matrix, axis=1)

# Display the results
for i, most_similar in enumerate(sim_articles):
    print(f"S {i+1}, S {most_similar+1}")
```
